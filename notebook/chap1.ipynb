## 1.2 因果推論の活用と価値創造

### Uber: クーポン配信

![alt text](images/chap1/image.png)

1. 無関心, 不干渉はクーポンを配布しても意味がない
2. 鉄板はクーポン配信しなくても利用してくれるのでコストの無駄
3. つまり説得可能群にクーポン配信するべき

説得可能群を特定するのに因果推論を使う。

顧客一人ひとりの売り上げ効果も推定できる。

クーポン配信時の売り上げ効果(因果効果) = 処置群(クーポンを配信した顧客)の平均売上 - 対照群 (クーポン配信なしの顧客)の平均売上

## 1.3 因果推論を活用した意思決定

意思決定に関わる要因や影響が膨大かつ、それぞれの要因がバイアスを含んでいるので、因果推論による精緻な手法論が必要

因果性が説明可能

サイモンが提唱した意思決定のプロセス
1. 情報活動: 意思決定に必要な情報を収集
2. 設計活動: 考えられる選択肢を列挙
3. 選択活動: 選択肢を評価し、決定

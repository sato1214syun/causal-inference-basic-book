相関関係と因果関係は混同されやすいが大きく異なる。店舗別の広告費と売上は正の相関があるが、両者に因果関係があるかをどう示すか。もしかしたら、広告費が高い店舗は、規模の大きいから広告費をかけやすく、規模に応じて売上が高いのかもしれない。

#### シンプルな方法 (第1章のヒュームの3条件)

1. 原因と結果は時空間的に隣接している -> 交絡因子(原因と結果の療法に影響する因子)が存在しない
2. 原因は結果に時間的に先行している -> 左記の通り原因が先行している
3. 原因と結果は常につながっている -> 両者に相関関係がある

#### 他の基準 (Bradford Hillの基準)

| 項目               | 説明                                                                                                   |
| ------------------ | ------------------------------------------------------------------------------------------------------ |
| 相関関係の強さ     | 原因Aと結果Bの間に強い相関関係がある                                                                   |
| 相関関係の一致性   | 相関関係の大きさが、対対象や、実証手法が違うなど様々な状況で一致する                                   |
| 相関関係の特異性 | 原因A**以外に想定される原因**と、結果Bの相関は高くない、また原因Aと結果B**以外の結果**の相関も高くない |
| 時間的な先行性     | 原因は結果に先行する                                                                                   |
| 量と反応関係の成立 | 原因Aの値が大きくなると、単調に結果Bの値も大きくなる                                                   |
| 妥当性             | AとBの因果関係が各分野のの知見に基づいてもっともらしい                                                 |
| 先行知見との整合性 | これまでの先行研究や知見と首尾一貫している                                                             |
| 実験による知見     | 実験研究による証拠がある                                                                               |
| 他の知見との類似性 | すでに確立している別の因果関係と類似した関係や構造を持つ                                               |

この基準は、修正されつつ広く使われる。ただしこれで完全に因果関係が示せるわけではなく。あくまでガイドライン。

重要なのは、因果関係を特定するために、
- プロセス
- メカニズム
- 統計的関連性

から考える必要があるということ。

これを考える具体的な方法として、ランダム化比較試験(Randomized Controlled Trial: RCT)がある。

## 2.2 ランダム化比較試験とは
### 2.2.1 あるべき検証法

何かの原因の影響を分析する際の、エビデンスのレベル

| レベル  | 説明                                                                                        |
| ------- | ------------------------------------------------------------------------------------------- |
| レベル1 | 処置群の前後比較、または処置群と対照群の処置後の1時点データのみを使用した分析(変数調整なし) |
| レベル2 | レベル1において変数調整を行った分析                                                         |
| レベル3 | 処置群と対照群それぞれの処置前後のデータを使用した分析(DiDなど)                             |
| レベル4 | 実験に類似する状況を利用した、回帰不連続デザインや操作変数法による分析                      |
| レベル5 | 処置群と対照群のランダム割り当てによる分析                                                  |

ランダム化比較試験(RCT)はレベル5に該当する。
- 究極的な目的: 完全に公平な条件下での因果効果の検証
- 理想: 特定のメカニズムについて、同一の対象に対して複数の異なる方法を同時に適用し、それぞれの方法がどれだけの効果があるかを評価したい
- 現実: 時間は絶えず進行するので、同一の対象に複数の方法を適用できない
- RCTではランダム割当によりほぼ同じ条件のグループ(群)を作り、処置の効果を検証する
- ランダム化をすると、観測されているかにかかわらず、交絡因子の影響を排除できる
- 例
    - ワクチンとプラセボ
    - web広告のA/Bテスト

ただしRCTは万能ではない

### 2.2.2 RCTの問題点

1. 倫理的にRCTが適用できない: ある群にのみ処置を行うことがはばかられる
2. 対象者が実験方法に従わない場合が存在
3. サンプルサイズが小さい場合。処置群と対照群に偏りが発生
4. 実施に費用と労力がかかる
5. 群同士を比較した平均的な結果が得られるため、サンプルの個別の因果効果を知りたい場合

などでRCTが使えない場合どうするか(次節以降で解説)

## 2.4 因果効果

### 2.4.1 因果効果とは -反実仮想で因果効果を考える-

理想的には、同一サンプル(個体)において、処置を実施した場合としなかった場合の結果を比較したい。この結果の差が因果効果。

あるサンプル$i$に対し、処置の有無($T=1, 0$)の結果($Y_i(1), Y_i(0)$)があるとする。

$|Y_i(1), Y_i(0)|$を潜在的結果変数と呼ぶ。

このアプローチを、処置を行ったかどうかの事実とは反対の結果を仮想することから、反実仮想アプローチと呼ばれる。実際はこの方法を実施することは不可能。

因果効果の対象を個体から集団に広げることで、統計的に因果効果を推定する事が可能になる。RCTもこの方法。

### 2.4.2 因果効果の種類(ATE、CATE、ITE)

因果効果の種類

#### ATEとATT 
- 平均処置効果 (Average Treatment Effect: ATE): 集団全体が処置を受けた場合の期待値と、受けなかった場合の期待値の差 $E(Y(1)) - E(Y(0))$
- 処置群への平均処置効果 (Average Treatment effect on the Treated: ATT): ATEを処置群のみに適用(ワクチンの副反応の効果など、処置群にしか起こらない事象について考えるとき使う)

#### CATEとITE
- 条件付き平均処置効果 (Conditional Average Treatment Effect: CATE): 集団を集団を何らかの条件でアブグループに分け、その中でATEを算出する
- 個別の因果効果 (Individual Treatment Effect：ITE): 個体別の因果効果を推定

施策を実施できる対象が限られているときに使う。予算の都合がある場合はより効果がある人を対象にしたいし、対象によってプラスの効果がでるかマイナスの効果が出るか変わる場合は、プラスになる人だけを判断して選びたい。

### 2.4.3 因果推論の重要な仮定 (識別性条件 + SUTVA)
- 識別性条件
    1. 独立性(交換可能性): 処置の有無にかかわらず、結果Yは独立
        
        ```
        独立性を満たさないケース
        例えば、ある感染症の重症患者にのみ、投薬を行い、軽症患者には投薬しない場合。
        - 投薬(処置あり)の群では、重症患者が多いので死亡率が高い
        - 投薬(処置なし)の群では、軽症患者が多いので死亡率が低い
        このようなケースは、処置(投薬)の有無と、結果(死亡)の間に薬の効果以外の相関があり独立性を満たさない

        独立性を満たすケース
        重症度と関係なく、投薬(処置)の有無を割り付ける場合(RCTを適用する)

        条件付き独立性
        上記の独立性を満たすケースでは、現実にはランダム割り付けしたグループ間で、重症患者と軽症患者の割合のバランスが取れず独立性が成立しないだろう。
        重症度が識別できるのであれば、重症度を共変量Xとして導入することで、処置群、対照群それぞれの中で重症/軽症を層別するとうまくATEが計算できる
        ```

        数式
        - 独立性を表す式
            - $\left\{Y(1),Y(0)\right\}\perp T$
        - 独立性を満たす場合以下の式が成り立つ。**つまり条件付け(処置の有無)は関係なくなる。**
            - $E[Y(1)|T=1]=E[Y(1)|T=0]=E[Y(1)]$
            - $E[Y(0)|T=0]=E[Y(0)|T=1]=E[Y(0)]$
        - なぜなら、条件付き期待値の定義は以下であり、
            - 離散値: $E[Y|T]=\sum yP(Y|T)$ 
            - 連続値: $E[Y|T]=\int yf_{Y|T}(y|t)dy$
        - 独立性を満たす(TとYが独立の)場合は以下となる
            - $P(Y|T)=P(Y)$
            - $f_{Y|T}(y|t)=f_Y(y)$
        - 条件付き独立性 (重症X=1, 軽症X=0とする)
            - $\{Y(0),Y(1)\}\bot T|X$
            - $E[Y(1)|X=1]=E[Y(1)|T=1,X=1]$
            - $E[Y(1)|X=0]=E[Y(1)|T=1,X=0]$

    2. 正値性: 各処置に割り当てられる確率が0ではない
       
       ある群において、確実に処置ありになる、もしくはなしになるということがないこと。
       
       共変量で条件付けをした場合は、条件付き正値性とよぶ
- SUTVA
    1. 相互作用なし
       
       ある個体への処置が、他の個体に影響を与えないこと
       
       Aにワクチンを打って、Bに打たず、AとB,それ以外の人で隔離すると、AとBが接触したときのBの感染リスクが減る(Aの状態に影響を受ける)
    2. 一致性
       
       処置を実際に受けたときの結果が、潜在的な結果と一致すること。処置の内容が曖昧な場合(例えば投薬量が決まっていないなど)はこれを満たさない

### 2.5.2 交絡因子・共変量

- 交絡因子: 背景因子として、原因にも結果にも影響を与えている因子
    - 早期の英語教育と総合学力の相関は、裏に家庭の収入という因子があるかも
    - そのままだと独立性が成り立たないので、交絡因子を調整して独立性を担保する必要がある
    - 交絡因子の調整方法は、バックドアパスを消すか、条件付け
- 共変量: 処置(原因)とは関係なく、結果に影響を及ぼす変数
- 中間変数: 処置(原因)と結果の因果の間に発生する変数

### 2.5.3 グラフィカルモデルを用いた因果の表現

グラフィカルモデル: グラフ理論のグラフで因果関係を表現したもの。非巡回グラフ(Directed Acyclic Graph: DAG)がよく用いられる。確率変数間の依存関係(因果の強さ)を加えればベイジアンネットワークになる

### 2.5.4 グラフィカルな識別

DAGによる因果関係の表し方

![図 2.25](images/chap2/image-4.png)

![図 2.26](images/chap2/image-5.png)

グラフの形がd分離を満たしていると、独立性があると判定できる
- d分離
    1. DAGの中でXとYを結ぶ任意の経路において、図2.25のような合流点Zやその子孫が頂点の部分集合Sに含まれていない
        1. つまり因果関係A,Bの途中に合流経路がある場合、Zを条件付けするとX,Yが独立から従属(交絡)に変化するので、独立性が崩れる(図2.26)
    2. XとYを結ぶ経路が連鎖経路や分岐経路であり、その中間ノードZが集合Sに含まれる
        1. Zを条件付けするとX,Yが独立になる(図2.26)

- バックドア基準(パックドアパスを閉じる条件)
    1. XからZへの有向道がない
    2. XとYの間でXに向かう有向道(バックドアパス)のすべてをZがd分離(ブロック)する

交絡因子は、バックドアを閉じるために必要な必要最低限の観測関数(条件付けする変数)の組み合わせと定義できる

この交絡因子を特定する操作をPythonのY-learnを用いて行うことができる

````python
from ylearn.causal_model.graph import CausalGraph
from ylearn.causal_model.model import CausalModel

causation = {
    "X1": [],
    "X2": [],
    "X3": ["X1"],
    "X4": ["X1", "X2"],
    "X5": ["X2"],
    "X6": ["X", "X1", "X2"],
    "X": ["X3", "X4", "X5"],
    "Y": ["X3", "X4", "X5", "X6", "X"],
}

cg = CausalGraph(causation=causation)
cm = CausalModel(causal_graph=cg)
backdoor_set, prob = cm.identify(
    treatment={"X"}, outcome={"Y"}, identify_method=("backdoor", "simple")
)["backdoor"]

print(backdoor_set)
````

````python
import matplotlib.pyplot as plt
import networkx as nx

##### 以下、グラフの可視化用コード ####

# Create directed graph
G = nx.DiGraph()

# Add nodes
for node in causation:
    G.add_node(node)

# Add edges based on causation
for target, sources in causation.items():
    for source in sources:
        G.add_edge(source, target)

# Draw the graph
pos = nx.spring_layout(G, seed=42)
labels = {node: node for node in G.nodes()}
nx.draw(
    G,
    pos,
    labels=labels,
    with_labels=True,
    node_color="skyblue",
    font_weight="bold",
    node_size=1000,
    font_size=15,
)
plt.show()
````

バックドア基準を満たすためには、少なくとも交絡因子がすべて観測されている必要がある。

未観測の変数が存在する場合でも因果効果を識別可能。その一つがフロントドア基準。

- フロントドア基準
    
    変数の集合Zが以下3つの条件を満たすとき、Zは変数X,Yについてフロントドア基準を満たす
    1. ZはXからYへの有向道をすべてブロックする
    2. XからZへのバックドアパスは存在しない
    3. ZからYへのすべてのバックドアパスはXによりブロックされている

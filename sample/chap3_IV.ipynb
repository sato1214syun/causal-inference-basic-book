```python
# 使用するライブラリのインポート
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.sandbox.regression.gmm import IV2SLS

# 再現性のためのシードを設定
np.random.seed(0)

# サンプルデータを生成
n = 1000  # 観測値の数
Z = np.random.normal(size=n)  # 操作変数
u = np.random.normal(size=n)  # 誤差項
T = Z + np.random.normal(size=n)  # 処置変数
Y = T + u  # 結果変数

# DataFrameを作成
df = pd.DataFrame({
    'Y': Y,
    'T': T,
    'Z': Z,
    'u': u
})

# 結果変数を定義
Y = df['Y']

# 処置変数を定義
T = sm.add_constant(df['T'])

# 操作変数を定義
Z = sm.add_constant(df['Z'])

# 操作変数法を適用
iv = IV2SLS(Y, T, Z).fit()

# 回帰のサマリーを出力
print(iv.summary())
```
